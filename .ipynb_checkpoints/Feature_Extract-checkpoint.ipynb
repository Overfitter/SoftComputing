```python
import numpy as np
import pyedflib
from nitime import utils
from nitime import algorithms as alg
from nitime.timeseries import TimeSeries
import csv
import pywt
import scipy.stats as sp
from os import listdir
from os.path import isfile, join
```

```python
names = ['max1','max2','max3','min1','min2','min3','Std1','Std2','Std3','Kurtosis1','Kurtosis2','Kurtosis3',
         'Skewness1','Skewness2','Skewness3','Result']
```

```python
def max_val(a):
    b = a #Extracting the data from the 3 channels
    output = np.zeros(len(b)) #Initializing the output array with zeros
    k = 0; #For counting the current row no.
    for l in b:
        i = l[~np.isnan(l)]
        output[k] = np.amax(i)#saving the max's
        k=k+1
    return output[0],output[1],output[2]
def min_val(a):
    b = a #Extracting the data from the 3 channels
    output = np.zeros(len(b)) #Initializing the output array with zeros
    k = 0; #For counting the current row no.
    for l in b:
        i = l[~np.isnan(l)]
        output[k] = np.amin(i)#saving the max's
        k=k+1
    return output[0],output[1],output[2]
def std_val(a):
    b = a #Extracting the data from the 3 channels
    output = np.zeros(len(b)) #Initializing the output array with zeros
    k = 0; #For counting the current row no.
    for l in b:
        i = l[~np.isnan(l)]
        output[k] = np.std(i) #saving the stds
        k += 1
    return output[0],output[1],output[2]
def my_kurtosis(a):
    b = a # Extracting the data from the 3 channels
    output = np.zeros(len(b)) # Initializing the output array with zeros (length = 3)
    k = 0; # For counting the current row no.
    for l in b:
        i = l[~np.isnan(l)]
        kurtosis_i = sp.kurtosis(i,fisher = True)
        output[k] = kurtosis_i # Saving the kurtosis in the array created
        k +=1 # Updating the current row no.
    return output[0],output[1],output[2]
def skewness(a):
    b = a # Extracting the data from the 3 channels
    output = np.zeros(len(b)) # Initializing the output array with zeros (length = 3)
    k = 0; # For counting the current row no.
    for l in b:
        i = l[~np.isnan(l)]
        skewness_i = sp.skew(i)
        output[k] = skewness_i # Saving the skewness in the array created
        k +=1 # Updating the current row no.
    return output[0],output[1],output[2]
```

```python
files9  = [f for f in listdir('DataCleaned/9') if isfile(join('DataCleaned/9', f))] 
files8  = [f for f in listdir('DataCleaned/8') if isfile(join('DataCleaned/8', f))]
files7  = [f for f in listdir('DataCleaned/7') if isfile(join('DataCleaned/7', f))] 
files6  = [f for f in listdir('DataCleaned/6') if isfile(join('DataCleaned/6', f))] 
files5  = [f for f in listdir('DataCleaned/5') if isfile(join('DataCleaned/5', f))]
files4  = [f for f in listdir('DataCleaned/4') if isfile(join('DataCleaned/4', f))] 
files3  = [f for f in listdir('DataCleaned/3') if isfile(join('DataCleaned/3', f))]
files2  = [f for f in listdir('DataCleaned/2') if isfile(join('DataCleaned/2', f))] 
files1  = [f for f in listdir('DataCleaned/1') if isfile(join('DataCleaned/1', f))]
files0  = [f for f in listdir('DataCleaned/0') if isfile(join('DataCleaned/0', f))]
files = []

for i in files9:
    files.append([i,'9'])
    
for i in files8:
    files.append([i,'8'])

for i in files7:
    files.append([i,'7'])

for i in files6:
    files.append([i,'6'])
    
for i in files5:
    files.append([i,'5'])

for i in files4:
    files.append([i,'4'])
    
for i in files3:
    files.append([i,'3'])

for i in files2:
    files.append([i,'2'])
    
for i in files1:
    files.append([i,'1'])
    
for i in files0:
    files.append([i,'0'])
    
read_path = 'DataCleaned/'
write_file_path = 'Features/features.csv'
write_file = open(write_file_path,"w")
writer = csv.writer(write_file, lineterminator='\n')
writer.writerow(names)


for counter in range(len(files)):
    subfolder =  files[counter][1]
    tag = files[counter][1]
    read_file_path = read_path + subfolder +'/'+files[counter][0]
    read_file = open(read_file_path,"rU")
    r = csv.reader(read_file,delimiter=',')
#     w = csv.reader(write_file,delimiter=',')
    data = np.empty((0,3), float)
    for l in r:
        data = np.append(data, np.array([l]), axis=0)
    data = data.astype(float)
    data = np.transpose(data)
    features = []
    
    max1,max2,max3 = max_val(data)
    features.append(max1)
    features.append(max2)
    features.append(max3)
    
    min1,min2,min3 = min_val(data)
    features.append(min1)
    features.append(min2)
    features.append(min3)
    
    Std1,Std2,Std3 = std_val(data)
    features.append(Std1)
    features.append(Std2)
    features.append(Std3)
    
    Kurtosis1,Kurtosis2,Kurtosis3 = my_kurtosis(data)
    features.append(Kurtosis1)
    features.append(Kurtosis2)
    features.append(Kurtosis3)
    
    Skewness1,Skewness2,Skewness3 = skewness(data)
    features.append(Skewness1)
    features.append(Skewness2)
    features.append(Skewness3)
    
    features.append(tag);
    writer.writerow(features)
    
write_file.close()
```

```python
r = csv.reader(open('Features/features.csv')) # Here your csv file
lines = [l for l in r]
```

```python
for i in range(len(lines[1])-1):  
    columns = []
    for j in range(1,len(lines)):
        columns.append(float(lines[j][i]))
    mean = np.mean(columns,axis = 0)
    std_dev  = np.std(columns,axis = 0)
    
    for j in range(1,len(lines)):
        lines[j][i] = (float(lines[j][i])-mean)/std_dev
write_file = open('Features/Normalizedfeatures.csv', 'wb')
writer = csv.writer(write_file) # This file will store the normalized features
writer.writerows(lines)
write_file.close()
```
